```python
c = widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre:',
    disabled=False,
)

def graph(a):
  dr = video[video['Genre']==a]
  gt = dr.groupby('channelTitle')
  gt = gt.sum().sort_values(by = 'view_count',ascending = False).head()
  sns.barplot(y = gt.index,x = gt['view_count'])
  plt.title(a)
  plt.xlabel('Views')
  plt.ylabel('Channel Name')
  plt.show()
```

```python
widgets.interact(graph,a=c)
```

```python
p = widgets.Dropdown(
    options=video['channelTitle'].unique(),
    value='jf.okay',
    description='channel:',
    disabled=False,
)

def channel(c):
  vc = video[video['channelTitle']==c]
  vc = vc.sort_values(by = 'view_count', ascending = False).head()
  sns.barplot(y = 'title' , x = 'view_count' , data = vc)
  plt.title(c)
  plt.show()
```

```python
widgets.interact(channel, c=p)
```

```python
g = widgets.Dropdown(
    options=video['Genre'].unique(),
    value='Music',
    description='Genre',
    disabled=False,
)

def genre(m):
  gen = video[video['Genre']==m]
  k2 = widgets.Dropdown(
    options=gen['channelTitle'].unique(),
    value=gen['channelTitle'].unique()[0],
    description='channel',
    disabled=False,
  )

  def channel(d):
    ht = gen.sort_values(by = 'view_count',ascending = False).head()
    sns.barplot(y = 'title',x = 'view_count', data = ht)
    plt.title(d)
    plt.xlabel('Views')
    plt.ylabel('Video Name')
    plt.show()

  widgets.interact(channel,d = k2)
```
